### We have to look at what kind of job people are most interested in so we can create our course and market accordingly and target the most amount of people.

```python
#we are only selecting people who answer the job role interest question so we have to remove the null
df.loc[df['JobRoleInterest'].notnull(), 'JobRoleInterest'].shape
df[df['JobRoleInterest'].notnull()]['JobRoleInterest']
```

From a quick scan we can see that most students are interested in web development

```python
#removing null and creating dictionary to hold counts
df_no_na = df['JobRoleInterest'].dropna()
roles = df_no_na.str.split(',')

total = {}
for i in roles:
    for r in i:
        o = r.lower().strip()
        if o in total:
            total[o] += 1
        else:
            total[o] = 1
```

### Lets look at top 20 jobs that people are interested in

```python
per_job = pd.DataFrame.from_dict(total, orient='index').rename(columns={0:'count'})
jobs = per_job['count'].sort_values(ascending= False)

top20_jobs = jobs[:20].copy()
others = (jobs[30:])
```

```python
top20_jobs
```

```python
top20_jobs.plot.bar(figsize = [8,5])
plt.ylabel('count')
plt.xlabel('Job Interest')
plt.title('Top 20 job interest')
```

#### And we saw that top 3 are in web development, follow by mobile and data science

### Also from the Survey, It's interesting to note that many respondant have more than one job interest so it would be useful to know how many job roles each person interested in

```python
roles_count = roles.apply(lambda x: len(x))
```

```python
absolute_count = roles_count.value_counts().sort_index()
relative_count = roles_count.value_counts(normalize = True).sort_index() * 100
```

```python
absolute_count
```

```python
df_job_count = pd.DataFrame({'absolute count' : absolute_count, 
                             'relative percentage' : relative_count})
df_job_count
```

#### so we know that about 31% respondant have clear idea of what job they are interested but rest of them have mixed idea and intrested in more than one  job

## How much does people spend on learning?

We will focus our analysis on US, India, UK, and Canada since 
- they are all english speaking
- all are highest frequency or top 4 country, below are about 1% each


We will focus on figuring out how much each student spend per month

```python
#we are replacing months learning to 1 so we won't get error when dividing
df_no_null_jobs['MonthsProgramming'].replace(0,1, inplace = True)
```

```python
#creating new column called "how much money student spend each month" and first let find out where they are null
df_no_null_jobs['money_spend_per_month'] = df_no_null_jobs['MoneyForLearning'] / df_no_null_jobs['MonthsProgramming']
df_no_null_jobs['money_spend_per_month'].isnull().sum()
```

```python
#removing all the null value from both money spend, and country they live since we need both column values
no_null_money = df_no_null_jobs[df_no_null_jobs['money_spend_per_month'].notnull()]
no_null_ct = no_null_money[no_null_money['CountryLive'].notnull()]

no_null_ct['CountryLive'].value_counts()[:5]
```

### Average Spending per country

```python
countries_mean = no_null_ct.groupby('CountryLive').mean()

countries_mean['money_spend_per_month'][['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

### Median Spending per country

```python
countries_median = no_null_ct.groupby('CountryLive').median()

countries_median['money_spend_per_month'][['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

```python
from matplotlib import rcParams

#figure size in inches
rcParams['figure.figsize'] = 11.7,8.27

top_4 = no_null_ct[no_null_ct['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

sns.distplot(top_4['money_spend_per_month'])
sns.plt.ylim(0, 0.00003)
```

##### The Result for average spending  shows that US has most amount spends and follow by India, UK and Canada.  This is surpising because we would assume UK and Canada to spend more than India, if we considser socio-economic metrics like GDP Per capita.

It might be there we don't have enough representative data.

From looking at the median it seems there are outlier because median is very small closer to 0 while mean is around 100s which could be from people attending bootcamps which cost thousands of dollars

## Let Visualize those Outlier and see if we can remove them and how would it affect our result

```python
# Isolate only the countries of interest
top_4 = no_null_ct[no_null_ct['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

from matplotlib import rcParams

#figure size in inches
rcParams['figure.figsize'] = 11.7,8.27
# Box plots to visualize distributions
import seaborn as sns
sns.boxplot(y = 'money_spend_per_month', x = 'CountryLive',
            data = top_4)
plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

It is hard to see the box plot because there are so many extreme outlier, especially in the US there is 2 person spending more than \$50000, this is not impossible but highly unlikely so let limit our data to below \$10000 per month 


```python
# Isolate only those participants who spend less than 10000 per month
no_null_less_10000 = no_null_ct[no_null_ct['money_spend_per_month'] < 10000]
```

```python
countries_mean = no_null_less_10000.groupby('CountryLive').mean()

countries_mean['money_spend_per_month'][['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

```python
# Isolate only the countries of interest
top_4 = no_null_less_10000[no_null_less_10000['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

# Box plots to visualize distributions
import seaborn as sns
sns.boxplot(y = 'money_spend_per_month', x = 'CountryLive',
            data = top_4)
plt.title('Money Spent Per Month Per Country',
         fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

We can still see outlier but now we can visualize closely and examine what those data point actually are

```python
india_outliers = top_4[
    (top_4['CountryLive'] == 'India') & 
    (top_4['money_spend_per_month'] >= 2500)]
india_outliers
```

It seems that any of the outlier doesn't attended bootcamp so it is unlikely that they are spending 5000 per month since the question in survey asked "Aside from university tuition, how much do you spend on learning code so far" so they may have misunderstood so we can remove these outlier

```python
# Remove the outliers for India
top_4 = top_4.drop(india_outliers.index) # using the row labels
```

```python
# Examine the extreme outliers for the US
us_outliers = top_4[
    (top_4['CountryLive'] == 'United States of America') & 
    (top_4['money_spend_per_month'] >= 6000)]

us_outliers
```

Here looking at the row, we will remove participants that had been programming for no more than two months when he completed the survey. This mean the person paid large sum of money in the bootcamp to enroll and then didn't spend anything for next couple of month therefore we should remove it.

```python

# Remove the respondents who didn't attendent a bootcamp
no_bootcamp = top_4[
    (top_4['CountryLive'] == 'United States of America') & 
    (top_4['money_spend_per_month'] >= 6000) &
    (top_4['AttendedBootcamp'] == 0)
]

top_4 = top_4.drop(no_bootcamp.index)


# Remove the respondents that had been programming for less than 3 months
less_than_3_months = top_4[
    (top_4['CountryLive'] == 'United States of America') & 
    (top_4['money_spend_per_month'] >= 6000) &
    (top_4['MonthsProgramming'] <= 3)
]

top_4 = top_4.drop(less_than_3_months.index)
```

```python
# Examine the extreme outliers for Canada
canada_outliers = top_4[
    (top_4['CountryLive'] == 'Canada') & 
    (top_4['money_spend_per_month'] > 4500)]

canada_outliers
```

We will also remove it because the situaltion is the same, attended bootcamp and never spend anymore after

```python

# Remove the extreme outliers for Canada
top_4 = top_4.drop(canada_outliers.index)
```

```python
# Recompute mean sum of money spent by students each month
top_4.groupby('CountryLive').mean()['money_spend_per_month']
```

```python
sns.boxplot(y = 'money_spend_per_month', x = 'CountryLive',
            data = top_4)
plt.title('Money Spent Per Month Per Country)',
          fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

```python
# Recompute mean sum of money spent by students each month
avg_top_4_money_spend_per_monthy = top_4.groupby('CountryLive').mean()['money_spend_per_month']
avg_top_4_money_spend_per_monthy
```
